```python
import pandas as pd
from IPython.display import display

df = pd.read_excel('OrderHistory.xlsx')
cleanDf = pd.DataFrame()
cleanFilledDf = pd.DataFrame()

# Clean the data
for index, row in df.iterrows():
    if row['Pair'].isalpha() and row['Pair'] != 'Date':
        cleanDf = cleanDf.append(row)
        if row['status'] == 'Filled':
            cleanFilledDf = cleanFilledDf.append(row)
        

groupDf = cleanDf.groupby('Pair')
cleanFilledDf = cleanFilledDf.groupby('Pair')

for name, group in cleanFilledDf:
    display(group)
```
